```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = {
    'Models': ['SVM', 'Decision Tree', 'Logistic Regression', 'NaÃ¯ve Bayes', 'Random Forest', 'AdaBoost', 'Gradient Boosting Classifier'],
    'Accuracy': [0.54, 0.7, 0.7, 0.54, 0.55, 0.55, 0.6],
    'Precision Score': [0.54, 0.72, 0.68, 0.54, 0.54, 0.54, 0.58],
    'Recall Score': [1, 0.71, 0.83, 1, 0.99, 0.99, 0.93],
    'F1 Score': [0.7, 0.72, 0.75, 0.7, 0.7, 0.7, 0.71]
}
df = pd.DataFrame(data)
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
df_melted = df.melt(id_vars='Models', var_name='Metric', value_name='Value')
sns.barplot(data=df_melted, x='Models', y='Value', hue='Metric')
plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Performance Metrics for Different Models')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Metrics', loc='upper right')
plt.tight_layout()
plt.show()
```
